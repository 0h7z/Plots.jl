```julia
pts = [
    (-2.5, 2.5),
    (-3.2, 1.5),
    (-0.5, -3.5),
    (0.5, -3.5),
    (3.2, 1.5),
    (2.5, 2.5),
    (-2.5, 2.5),
    (-0.5, -1.5),
    (0.0, -0.5),
    (-1.0, 1.5),
    (3.2, 1.5),
    (0.5, -3.5),
    (-0.5, -3.5),
    (1.5, 0.5), 
    (-0.5, 0.5),
    (0.0, -0.5),
    (0.5, 0.5)
]

scalar = 1 / 2.5
pts = [(x*scalar, y*scalar) for (x,y) in pts]

x, y = Float64[x for (x,y) in pts], Float64[y for (x,y) in pts]

plot(x,-y)
```
